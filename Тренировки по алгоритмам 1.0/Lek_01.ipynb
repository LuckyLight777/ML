## 1. Найти элемент, который встречается чаще всего

Дана строка. Найти самый часто встречающийся в ней символ. Если несколько символов встречаются одинаково часто, вывести любой символ.

**Решение 1** \\
Время: $O(N^2)$ \\
Память: $O(N)$

```python
# s = input()
s = '73777335447'
ans = ''
kol_max = 0
for i in range(len(s)):
  kol = 0
  for j in range(len(s)):
    if s[i] == s[j]:
      kol += 1
  if kol > kol_max:
    ans = s[i]
    kol_max = kol
print(ans)
```

**Решение 2** \\
Время: $O(k*N)$\\
Память: $O(N+k)

```python
# s = input()
s = '73777335447'
ans = ''
kol_max = 0
for symb in set(s):
  kol = 0
  for i in range(len(s)):
    if symb== s[i]:
      kol += 1
  if kol > kol_max:
    ans = symb
    kol_max = kol
print(ans)
```

**Решение 3** \\
Сложность: $O(k+N)$

```python
# s = input()
s = '7373337733544733333333'
d = {}
for i in s:
  if i not in d:
    d[i] = 0
  d[i] += 1

kol_max = 0
for key in d:
  if d[key] > kol_max:
    kol_max = d[key]
    ans = key

print(ans)
```

```python
# s = input().split()
s = "-7 -2 -3 -4 -5".split()

def max_seq(s):
  seq = list(map(int, s))
  if len(seq) == 0:
    return ('-inf')
  max_seq = seq[0]
  for i in range(1, len(seq)):
    if max_seq < seq[i]:
      max_seq = seq[i]
  return max_seq

print(max_seq(s))
```

## 5. Скорая помощь

Бригада скорой помощи выехала по вызову в один из отделенных районов. К сожалению, когда диспетчер получил вызов, он успел записать только адрес дома и номер квартиры K1, а затем связь прервалась. Однако он вспомнил, что по этому же адресу дома некоторое время назад скорая помощь выезжала в квартиру K2, которая расположена в подъезда P2 на этаже N2. Известно, что в доме M этажей и количество квартир на каждой лестничной площадке одинаково. Напишите программу, которая вычисляет номер подъезда P1 и номер этажа N1 квартиры K1.

```python
def loc_flat(k, k_n, m):
    p = (((k - 1) // k_n) // m) + 1
    n = (((k - 1) // k_n) % m) + 1
    return p, n

def check(k1, m, k2, p2, n2, k_n):
    p_2, n_2 = loc_flat(k2, k_n, m)
    if p_2 == p2 and n_2 == n2:
        return loc_flat(k1, k_n, m)
    return -1, -1

def flats_on_floor(k1, m, k2, p2, n2):
    p1 = -1
    n1 = -1
    flag = False

    for k_n in range(1, max(k1, k2) + 1):
        p, n = check(k1, m, k2, p2, n2, k_n)
        if p != -1:
            flag = True
            if p1 == -1:
                p1, n1 = p, n
            else:
                if p1 != p:
                    p1 = 0
                if n1 != n:
                    n1 = 0

    if flag:
        return p1, n1
    else:
        return -1, -1

k1, m, k2, p2, n2 = map(int, input().split())

p1, n1 = flats_on_floor(k1, m, k2, p2, n2)
print(p1, n1)
```

## 7. Детали

Имеется N кг металлического сплава. Из него изготавливают заготовки массой K кг каждая. После этого из каждой заготовки вытачиваются детали массой M кг каждая (из каждой заготовки вытачивают максимально возможное количество деталей). Если от заготовок после этого что-то остается, то этот материал возвращают к началу производственного цикла и сплавляют с тем, что осталось при изготовлении заготовок. Если того сплава, который получился, достаточно для изготовления хотя бы одной заготовки, то из него снова изготавливают заготовки, из них – детали и т.д. Напишите программу, которая вычислит, какое количество деталей может быть получено по этой технологии из имеющихся исходно N кг сплава.

```python
N, K, M = map(int, input().split())

m = 0
if K >= M:
  while N >= K:
    k = N // K                #количество заготовок
    m += K // M * k           #количество деталей
    N = N % K + K % M * k     #остаток сплава
print(m)
```
