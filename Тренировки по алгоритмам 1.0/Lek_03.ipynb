**Хеш-функция** — функция, преобразующая массив входных данных произвольного размера в выходную битовую строку определённого (установленного) размера в соответствии с определённым алгоритмом. Преобразование, выполняемое хеш-функцией, называется хешированием.
(F(x) = x % setsize)

**Хеш-таблица** — структура данных, реализующая интерфейс ассоциативного массива, а именно, она позволяет хранить пары (ключ, значение) и выполнять три операции: операцию добавления новой пары, операцию удаления и операцию поиска пары по ключу.
(my_set - список списков)


**Коллизия** - совпадение значений хэш-функции для различных параметров

```python
import random as rnd

class my_set:
  def __init__(self, l = [], rand = False):
    self.setsize = 10
    self.my_set = [[] for _ in range(self.setsize)]
    if rand:
      self.rand_create(rand)
    if l:
      for i in l:
        self.add(i)

  def rand_create(self, size):
    for i in range(size):
      self.add(rnd.randint(0, 100))

  def output(self):
    flag = False
    for i in self.my_set:
      if i:
        # print(i)
        flag = True
        for j in i:
          print(j, end=' ')
    print()
    if not flag:
      print('Empty')
    print()

  def check_extension(self):
    for sp in self.my_set:
      if not sp:
        return False
    return True

  def extension(self):
    self.setsize *= 2
    new_set = [[] for _ in range(self.setsize)]
    for sp in self.my_set:
      for i in sp:
        new_set[i % self.setsize].append(i)
    self.my_set = new_set

  def add(self, x):
    if self.find(x):
      return
    self.my_set[x % self.setsize].append(x)
    if self.check_extension():
      self.extension()

  def find(self, x):
    for i in self.my_set[x % self.setsize]:
      if i == x:
        return True
    return False

  def delit(self, x):
    sp = self.my_set[x % self.setsize]
    for i in range(len (sp)):
      if sp[i] == x:
        sp[i] = sp[len (sp) - 1]
        sp.pop()
        return


# s = my_set(rand = 15)
s = my_set([1, 2, 33, 84, 185, 386, 7, 83, 29, 10])
s.output()
s.add(58)
s.add(22)
s.output()
```

Дана последовательность положительных чисел длинной N. Нужно найти 2 числа А и В, таких что А + В = Х или вернуть 0, 0, если таких чисел нет

## Задача 2: Проверка слов в словаре

Дан словарь из N слов. Длина каждого не превосходит K. В записи каждого из M слов текста может быть пропущена одна буква. Для каждого слова сказать, входит ли оно в словарь.

```python
dic = {'a', 'bbbb', 'cfhfj'}
text = 'a bbbb chfj'

def word_in_dic(dic, text):
  s = set(dic)
  for word in dic:
    for propuck in range(len(word)):
      s.add(word[:propuck] + word[propuck + 1:])

  ans = []
  for word in text.split():
    if word in s:
      ans.append(word)
  return ans

print(word_in_dic(dic, text))
```

## C. Кубики

Аня и Боря любят играть в разноцветные кубики, причем у каждого из них свой набор и в каждом наборе все кубики различны по цвету. Однажды дети заинтересовались, сколько существуют цветов таких, что кубики каждого цвета присутствуют в обоих наборах. Для этого они занумеровали все цвета случайными числами. На этом их энтузиазм иссяк, поэтому вам предлагается помочь им в оставшейся части. Номер любого цвета — это целое число в пределах от 0 до 109.

```python
N, M = map(int, input().split())

a = set()
for i in range(N):
  a.add(int(input()))

b = set()
for i in range(M):
  b.add(int(input()))

print(len(a & b))
print(*sorted(a & b))
print(len(a - b))
print(*sorted(a - b))
print(len(b - a))
print(*sorted(b - a))
```

## F. Инопланетный геном

Геном жителей системы Тау Кита содержит 26 видов оснований, для обозначения которых будем использовать буквы латинского алфавита от A до Z, а сам геном записывается строкой из латинских букв. Важную роль в геноме играют пары соседних оснований, например, в геноме «ABBACAB» можно выделить следующие пары оснований: AB, BB, BA, AC, CA, AB.

Степенью близости одного генома другому геному называется количество пар соседних оснований первого генома, которые встречаются во втором геноме.

```python
def add(my_set, x):
    my_set[ord(x[0]) - ord('A')].append(x)

line1 = input()
line2 = input()

setsize = 26
gen1 = [[] for _ in range(setsize)]
for i in range(len(line1) - 1):
  add(gen1, line1[i] + line1[i + 1])

gen2 = [[] for _ in range(setsize)]
for i in range(len(line2) - 1):
  add(gen2, line2[i] + line2[i + 1])

count = 0
for i in range(len(gen1)):
  for j in gen1[i]:
    if j in gen2[i]:
      count += 1
print(count)
```

## G. Черепахи

Широко известна следующая задача для младших школьников. Три черепахи ползут по дороге. Одна черепаха говорит: “Впереди меня две черепахи”. Другая черепаха говорит: “Позади меня две черепахи”. Третья черепаха говорит: “Впереди меня две черепахи и позади меня две черепахи”. Как такое может быть? Ответ: третья черепаха врет! По дороге одна за другой движутся N черепах. Каждая черепаха говорит фразу вида: “Впереди меня ai черепах, а позади меня bi черепах”. Ваша задача определить, сколько самое большее количество черепах могут говорить правду.

```python
n = int(input())

s_a = set()
s_b = set()
for i in range(n):
  ai, bi = map(int, input().split())
  if ai + bi + 1 == n:
    s_a.add(ai)
    s_b.add(bi)
print(min(len(s_a), len(s_b)))
```
