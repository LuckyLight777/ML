# –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫

```python
def search(arr: list, value: int) -> bool:
  if not arr:
    return False

  a = 0
  b = len(arr) - 1
  while a <= b:
    mid = (a + b) // 2
    if arr[mid] == value:
      return True
    elif arr[mid] < value:
      a = mid + 1
    else:
      b = mid - 1
  return False

arr = [1, 2, 3, 4, 5]
val = 2
print(search(arr, val))
```

# –ù–∞–π—Ç–∏ —ç–ª–µ–º–µ–Ω—Ç, –∫–æ—Ç–æ—Ä—ã–π –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —á–∞—â–µ –≤—Å–µ–≥–æ

**–†–µ—à–µ–Ω–∏–µ 1** \\
–í—Ä–µ–º—è: $O(N^2)$ \\
–ü–∞–º—è—Ç—å: $O(N)$

```python
# s = input()
s = '73777335447'
ans = ''
kol_max = 0
for i in range(len(s)):
  kol = 0
  for j in range(len(s)):
    if s[i] == s[j]:
      kol += 1
  if kol > kol_max:
    ans = s[i]
    kol_max = kol
print(ans)
```

**–†–µ—à–µ–Ω–∏–µ 2** \\
–í—Ä–µ–º—è: $O(k*N)$\\
–ü–∞–º—è—Ç—å: $O(N+k)

```python
# s = input()
s = '73777335447'
ans = ''
kol_max = 0
for symb in set(s):
  kol = 0
  for i in range(len(s)):
    if symb== s[i]:
      kol += 1
  if kol > kol_max:
    ans = symb
    kol_max = kol
print(ans)
```

**–†–µ—à–µ–Ω–∏–µ 3** \\
–°–ª–æ–∂–Ω–æ—Å—Ç—å: $O(k+N)$

```python
# s = input()
s = '7373337733544733333333'
d = {}
for i in s:
  if i not in d:
    d[i] = 0
  d[i] += 1

kol_max = 0
for key in d:
  if d[key] > kol_max:
    kol_max = d[key]
    ans = key

print(ans)
```
