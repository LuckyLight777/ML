# Найти 2 элемента последовательности, чья сумма ближе всего к K

```python
k = int(input())
seq = list(map(int, input().split()))

def sum_k(seq, k):
  seq.sort()
  a = 0
  b = len(seq) - 1
  min_r = abs(seq[a] + seq[b] - k)
  min_a = a
  min_b = b
  while a < b:
    if min_r == 0:
      break
    if min_r > abs(seq[a] + seq[b] - k):
      min_r = abs(seq[a] + seq[b] - k)
      min_a = a
      min_b = b
    if k > seq[a] + seq[b]:
      a += 1
    else:
      b -= 1

  return seq[min_a], seq[min_b]

print(*sum_k(seq, k))
```
